#Intro to Pandas

Pandas is an open-source data manipulation and analysis library for Python.

It provides easy-to-use data structures and data analysis tools that make working with structured data, such as spreadsheets or SQL tables, efficient and intuitive.

Pandas is an essential library in the data science and data analysis ecosystem, and it is widely used by data professionals,researchers, and analysts for tasks such as data cleaning , exploration , transformation , and visualization.





##A  detailed description of Pandas:

1) Data Structures:
Pandas primarily offers two main data structures: Series and DataFrame.

Series: A Series is a one-dimensional labeled array capable of holding any data type. It is similar to a column in a spreadsheet or a single variable in statistics. Each element in a Series has a unique label or index.

DataFrame: A DataFrame is a two-dimensional, size-mutable, and heterogeneous tabular data structure. It consists of rows and columns, where each column can have a different data type. You can think of a DataFrame as a spreadsheet or a SQL table.

2) Data Input/Output:
Pandas provides functions to read data from various file formats such as CSV, Excel, SQL databases, and more. It also allows you to write data to these formats. This makes it easy to import and export data for analysis.

3) Data Cleaning and Preparation:
One of the key strengths of Pandas is its ability to handle missing data and data cleaning. It provides methods for filling in missing values, dropping rows or columns with missing data, and transforming data.

4) Data Selection and Indexing:
Pandas allows for flexible and powerful indexing and selection of data. You can select specific rows and columns, filter data based on conditions, and perform operations on selected subsets of data.

5) Data Transformation:
Pandas provides tools for reshaping and transforming data, such as merging, pivoting, melting, and grouping. These operations are crucial for preparing data for analysis and visualization.

6) Data Analysis and Exploration:
You can perform various statistical and analytical operations on data using Pandas. It includes aggregation, descriptive statistics, correlation analysis, and more. Pandas also integrates well with other data analysis libraries like NumPy, Matplotlib, and Seaborn for advanced visualization and analysis.

7) Time Series Data:
Pandas has robust support for working with time series data, making it a popular choice for financial analysis, IoT data analysis, and other applications involving temporal data.

8) Integration with Other Libraries:
Pandas can be seamlessly integrated with other Python libraries such as NumPy, Matplotlib, SciPy, and scikit-learn, allowing you to create comprehensive data analysis and machine learning pipelines.

9) Performance and Efficiency:
While Pandas is efficient for most data manipulation tasks, it may not be the best choice for extremely large datasets due to memory constraints. In such cases, you might need to use alternative tools like Dask or work with databases directly.

10) Community and Ecosystem:
Pandas has a large and active community of users and developers. This means that you can find a wealth of online resources, tutorials, and third-party packages that extend Pandas' functionality.

Pandas is a powerful and versatile library for data manipulation and analysis in Python.

It simplifies many common data-related tasks and is an essential tool in the toolkit of data scientists , analysts , and researchers.

It's flexibility and rich functionality make it a go-to choice for working with structured data in a wide range of applications.

#pandas.DataFrame.dropna

DataFrame.dropna(*, axis=0, how=_NoDefault.no_default, thresh=_NoDefault.no_default, subset=None, inplace=False, ignore_index=False)

Parameters:

***axis*** : ***{0 or ‘index’, 1 or ‘columns’}, default 0***
Determine if rows or columns which contain missing values are removed.

1) 0, or ‘index’ : Drop rows which contain missing values.

2) 1, or ‘columns’ : Drop columns which contain missing value.

Only a single axis is allowed.

***how*** :*** {‘any’, ‘all’}, default ‘any’***
Determine if row or column is removed from DataFrame, when we have at least one NA or all NA.

1)‘*any*’ : If any NA values are present, drop that row or column.

2)‘*all*’ : If all values are NA, drop that row or column.

***thresh*** ***: int, optional***
Require that many non-NA values. Cannot be combined with how.

***subset : column label or sequence of labels, optional***
Labels along other axis to consider, e.g. if you are dropping rows these would be a list of columns to include.

***inplace: bool, default False***
Whether to modify the DataFrame rather than creating a new one.

ignore_index :bool, default False ***bold text***
If True, the resulting axis will be labeled 0, 1, …, n - 1.

#Grouping
By “**group by**” we are referring to a process involving one or more of the following steps:

1) **Splitting** the data into groups based on some criteria

2) **Applying** a function to each group independently

3) **Combining** the results into a data structure

```python
df = pd.DataFrame(
    {
        "A" : ["foo","bar","foo","bar","foo","bar","foo","foo"],
        "B" : ["one","one","two","three","two","two","one","three"],
        "C" : np.random.randn(8),
        "D" : np.random.randn(8),
    }
)
```

```python
df
```

Grouping by a column label, selecting column labels, and then applying the **sum()** function to the resulting groups:

```python
df.groupby("A")[["C","D"]].sum()
```

Grouping by multiple columns label forms **MultiIndex**.

```python
df.groupby(["A","B"]).sum()
```

#Categoricals
pandas can include categorical data in a DataFrame.

```python
df = pd.DataFrame(
    {"id": [1, 2, 3, 4, 5, 6], "raw_grade": ["a", "b", "b", "a", "a", "e"]}
)
```

Converting the raw grades to a categorical data type:

```python
df["grade"] = df["raw_grade"].astype("category")

df["grade"]
```

Rename the categories to more meaningful names:

```python
new_categories = ["very good", "good", "very bad"]

df["grade"] = df["grade"].cat.rename_categories(new_categories)
```

Reorder the categories and simultaneously add the missing categories (methods under **Series.cat()** return a new  **Series** by default):

```python
df["grade"] = df["grade"].cat.set_categories(
    ["very bad", "bad", "medium", "good", "very good"])
```

```python
df["grade"]
```

Sorting is per order in the categories, not lexical order:

```python
df.sort_values(by="grade")
```

Grouping by a categorical column with ***observed=False*** also shows empty categories:

```python
df.groupby("grade", observed=False).size()
```
